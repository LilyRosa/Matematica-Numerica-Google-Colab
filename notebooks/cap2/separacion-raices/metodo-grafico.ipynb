```python
from sympy import symbols
from sympy.plotting import plot
import math

x = symbols('x')
```

# Método Gráfico

## Descripción:
Método que permite determinar las raíces de una ecuación algebraica o trascendente de forma visual gracias a su representación gráfica.

## Versión 1

### Procedimiento:
- Transformar la ecuación a la forma $f(x)=0$
- Tomar $f(x)$ como `f`
- Elegir un intervalo primario (Ej: $[-5,5]$) para reconocer las raíces de la ecuación, que es donde la función graficada corta el eje x
- Disminuir el intervalo dado para determinar el rango de cada raíz de la ecuación

```python
# Datos 
f = (math.e**-x)-x

# Graficado
plot(f, (x, 0, 1))
```

## Versión 2

### Procedimiento:
- Transformar la ecuación a la forma $f_1(x)=f_2(x)$
- Tomar $f_1(x)$ como `f1` y $f_2(x)$ como `f2`
- Elegir un intervalo primario (Ej: $[-5,5]$) para reconocer las raíces de la ecuación, que es donde ambas funciones se cortan
- Disminuir el intervalo dado para determinar el rango de cada raíz de la ecuación

```python
# Datos 
f1 = math.e**-x
f2 = x

# Graficado
plot(f1, f2, (x, 0, 1))
```
